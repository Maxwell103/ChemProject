```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Create a dataframe with coordinates of the triangle vertices
data = {'X': [0, 1, 2], 'Y': [0, 3, 0]}
df = pd.DataFrame(data)

# Plot the triangle
plt.plot(df['X'], df['Y'])
plt.fill(df['X'], df['Y'], alpha=0.5)  # Fill the triangle with color

# Generate random coordinates for the two lines touching the bottom corners
x1 = random.uniform(0, 1)  # Random x-coordinate for the first line
x2 = random.uniform(1, 2)  # Random x-coordinate for the second line
y1 = 0  # y-coordinate for the first line (bottom of the triangle)
y2 = 0  # y-coordinate for the second line (bottom of the triangle)

# Calculate the slope and intercept of the lines
m1 = 3 / x1  # Slope of the first line
b1 = y1  # Intercept of the first line
m2 = -3 / (x2 - 2)  # Slope of the second line
b2 = y2  # Intercept of the second line

# Calculate the intersection point
x_center = (b2 - b1) / (m1 - m2)
y_center = m1 * x_center + b1

# Check if the intersection point is within the bounds of the triangle
within_bounds = 0 <= x_center <= 2 and 0 <= y_center <= 3

# Plot the center lines if the intersection point is within the bounds
if within_bounds:
    plt.plot([x1, x_center], [y1, y_center], color='red')
    plt.plot([x2, x_center], [y2, y_center], color='blue')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triangle with Center Lines')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe with coordinates of the triangle vertices
data = {'X': [0, 1, 2], 'Y': [0, 3, 0]}
df = pd.DataFrame(data)

# Plot the triangle
plt.plot(df['X'], df['Y'])
plt.fill(df['X'], df['Y'], alpha=0.5)  # Fill the triangle with color

# Calculate the center of the triangle
center_x = df['X'].mean()
center_y = df['Y'].mean()

# Store the coordinates where the lines touch the triangle's edges
line_coords = []

# Generate curved lines and store the touch points
for i in range(3):
    if df['Y'][i] > 0:
        x = np.linspace(df['X'][i], center_x, 100)  # X-values for the curve
        cpx = (df['X'][i] + center_x) / 2  # X-value for the control point
        cpy = df['Y'][i]  # Y-value for the control point
        y = ((x - df['X'][i]) * (x - center_x) / (cpx - df['X'][i]) / (cpx - center_x)) * (cpy - df['Y'][i]) + df['Y'][i]  # Y-values for the curve

        plt.plot(x, y, color='red')
        touch_point = (x[-1], y[-1])
        line_coords.append(touch_point)
    else:
        line_coords.append((df['X'][i], df['Y'][i]))

# Plot the center point
plt.plot(center_x, center_y, 'ko', label='Center')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triangle with Curved Lines')
plt.legend()
plt.show()

# Print the touch points
print("Coordinates where the lines touch the triangle's edges:")
for i, coord in enumerate(line_coords):
    print(f"Line {i+1}: {coord}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe with coordinates of the triangle vertices
data = {'X': [0, 1, 2, 0], 'Y': [0, 3, 0, 0]}  # Include the first vertex to close the triangle
df = pd.DataFrame(data)

# Plot the triangle
plt.plot(df['X'], df['Y'])
plt.fill(df['X'], df['Y'], alpha=0.5)  # Fill the triangle with color

# Calculate the center of the triangle
center_x = df['X'].mean()
center_y = df['Y'].mean()

# Store the coordinates where the lines touch the triangle's edges
line_coords = []

# Define the label offset and initial step size for avoiding overlap
offset = 0.1
step = 0.05

# Plot the lines and store the touch points
for i in range(3):
    line_x = [df['X'][i], center_x]
    line_y = [df['Y'][i], center_y]
    plt.plot(line_x, line_y, color='red')

    touch_point = (line_x[1], line_y[1])
    line_coords.append(touch_point)

    # Compute the label position slightly away from the center
    label_x = line_x[1] + (center_x - line_x[1]) * offset
    label_y = line_y[1] + (center_y - line_y[1]) * offset

    # Check for label overlap and adjust position if necessary
    for j in range(i):
        # Compute the distance between labels
        dist = ((label_x - plt.texts[j].get_position()[0])**2 + (label_y - plt.texts[j].get_position()[1])**2)**0.5

        if dist < 0.1:
            # Increase the offset if the labels are too close
            offset += step
            label_x = line_x[1] + (center_x - line_x[1]) * offset
            label_y = line_y[1] + (center_y - line_y[1]) * offset

            # Reset the loop to check the adjusted position against previous labels
            j = -1

    # Label the lines
    plt.text(label_x, label_y, f'Line {chr(97+i)}', ha='center', va='bottom')

# Plot the center point
plt.plot(center_x, center_y, 'ko', label='Center')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triangle with Center Lines')
plt.legend()
plt.show()

# Print the touch points
print("Coordinates where the lines touch the triangle's edges:")
for i, coord in enumerate(line_coords):
    print(f"Line {chr(97+i)}: {coord}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a dataframe with coordinates of the triangle vertices
data = {'X': [0, 1, 2, 0], 'Y': [0, 3, 0, 0]}  # Include the first vertex to close the triangle
df = pd.DataFrame(data)

# Plot the triangle
plt.plot(df['X'], df['Y'])
plt.fill(df['X'], df['Y'], alpha=0.5)  # Fill the triangle with color

# Calculate the center of the triangle
center_x = df['X'].mean()
center_y = df['Y'].mean()

# Store the coordinates where the lines touch the triangle's edges
line_coords = []

# Plot the lines and store the touch points
for i in range(3):
    # Compute the control point for the quadratic curve
    control_x = (df['X'][i] + center_x) / 2
    control_y = (df['Y'][i] + center_y) / 2
    
    # Create an array of x-values between the two end points
    curve_x = np.linspace(df['X'][i], center_x, num=100)
    
    # Compute the corresponding y-values using the quadratic equation
    curve_y = df['Y'][i] + (curve_x - df['X'][i]) ** 2 * (control_y - df['Y'][i]) / (center_x - df['X'][i]) ** 2

    # Plot the curved line
    plt.plot(curve_x, curve_y, color='red')
    
    # Store the touch point
    touch_point = (curve_x[-1], curve_y[-1])
    line_coords.append(touch_point)

    # Label the lines
    plt.text(touch_point[0], touch_point[1], f'Line {chr(97+i)}', ha='center', va='bottom')

# Plot the center point
plt.plot(center_x, center_y, 'ko', label='Center')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triangle with Curved Center Lines')
plt.legend()
plt.show()

# Print the touch points
print("Coordinates where the lines touch the triangle's edges:")
for i, coord in enumerate(line_coords):
    print(f"Line {chr(97+i)}: {coord}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# Create a dataframe with coordinates of the triangle vertices
data = {'X': [0, 1, 2, 0], 'Y': [0, 3, 0, 0]}  # Include the first vertex to close the triangle
df = pd.DataFrame(data)

# Plot the triangle
plt.plot(df['X'], df['Y'])
plt.fill(df['X'], df['Y'], alpha=0.5)  # Fill the triangle with color

# Calculate the center of the triangle
center_x = df['X'].mean()
center_y = df['Y'].mean()

# Store the coordinates where the lines touch the triangle's edges
line_coords = []

# Plot the lines and store the touch points
for i in range(3):
    line_x = [df['X'][i], center_x]
    line_y = [df['Y'][i], center_y]
    plt.plot(line_x, line_y, color='red')

    touch_point = (line_x[1], line_y[1])
    line_coords.append(touch_point)

    # Label the lines
    label = f'Line {chr(97+i)}'
    ab = AnnotationBbox(OffsetImage(plt.imread(label), zoom=0.05), (line_x[1], line_y[1]), frameon=False)
    plt.gca().add_artist(ab)

# Plot the center point
plt.plot(center_x, center_y, 'ko', label='Center')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Triangle with Center Lines')
plt.legend()
plt.show()

# Print the touch points
print("Coordinates where the lines touch the triangle's edges:")
for i, coord in enumerate(line_coords):
    print(f"Line {chr(97+i)}: {coord}")
```
